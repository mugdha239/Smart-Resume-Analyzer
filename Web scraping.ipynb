```python

from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time
import pandas as pd
import os

# Set up the ChromeDriver service
service = Service(ChromeDriverManager().install())

# Initialize the Chrome WebDriver
driver = webdriver.Chrome(service=service)
```

```python
url1='https://www.linkedin.com/jobs/search?keywords=Marketing%20Data%20Analyst&location=Berlin&geoId=106967730&trk=public_jobs_jobs-search-bar_search-submit&position=1&pageNum=0'
```

```python
driver.get(url1)
```

```python
driver.implicitly_wait(10)
```

```python
# n=driver.find_elements_by_class_name('results-context-header__job-count')[0].text
# n
from selenium.webdriver.common.by import By

# Find elements using the updated method
elements = driver.find_elements(By.CLASS_NAME, 'results-context-header__job-count')

# Get the text of the first element
n = elements[0].text
n
```

```python
y=pd.to_numeric(n)
y
```

```python
i=2
while(i<=14):
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight);")
    i+=1
    
    try:
        x=driver.find_element_by_xpath("//button[@button aria-label='See more jobs']")
        driver.execute_script("arguments[0].click();",x)
        time.sleep(3)
    except:
        pass
        time.sleep(4)
```

```python
companyname=[]
```

```python
# for j in range(y):
#     try:
#         company=driver.find_elements_by_class_name('base-search-card__subtitle')[j].text
#         companyname.append(company)
#     except IndexError:
#         print("done")
        
for j in range(y):
    try:
        # Use the updated method to find elements by class name
        company = driver.find_elements(By.CLASS_NAME, 'base-search-card__subtitle')[j].text
        companyname.append(company)
    except IndexError:
        # Handle case where index is out of range
        print("done")
        break  # Exit loop if no more elements are found
    except Exception as e:
        # Handle any other exceptions
        print(f"An error occurred: {e}")

# Print or use the company names as needed
# print(companyname)
```

```python
companyname
```

```python
titlename=[]
for j in range(y):
    try:
        # Use the updated method to find elements by class name
        title = driver.find_elements(By.CLASS_NAME, 'base-search-card__title')[j].text
        titlename.append(title)
    except IndexError:
        # Handle case where index is out of range
        print("done")
        break  # Exit loop if no more elements are found
    except Exception as e:
        # Handle any other exceptions
        print(f"An error occurred: {e}")
titlename
```

```python
companyfinal=pd.DataFrame(companyname,columns=["Company"])
titlefinal=pd.DataFrame(titlename,columns=["Title"])
final=companyfinal.join(titlefinal)
final
```

```python
final.to_csv('linkedinjobdetails.csv')
```

```python
from selenium.webdriver.common.by import By

# Initialize an empty list to store links
linklist = []

# Find elements using the updated method
findlink = driver.find_elements(By.CLASS_NAME, 'base-card__full-link')

# Iterate over the elements and get their 'href' attribute
for k in findlink:
    linklist.append(k.get_attribute('href'))

# Print or use the linklist as needed
linklist
```

```python
linkfinal=pd.DataFrame(linklist,columns=["Link"])
```

```python
final1=final.join(linkfinal)
final1
```

```python
final1.to_csv('linkedinjobdetails1.csv')
```
